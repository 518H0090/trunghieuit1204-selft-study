```python
text = 'Text in python is called String'
text_2 = "Snake case ?"
num_int = 10
num_float = 3.14
boolean = True
none_type = None

print(text, text_2, num_int, num_float, boolean, none_type)
print(type(none_type))
print(type(num_float))
print(type(boolean))
```

```python
user = "Hello"
app  = "New"

print('Once upon a time, there was a user named ' + user)
print('My name is : {}'.format(user))
print(f'Once upon a time, there was a {app} user named {user}')
```
